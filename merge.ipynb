# IMPORT

```python
# dados de 2022
# 2022 data

first_path2022 = "C:\\Users\\SAMSUNG\\OneDrive\\Documentos\\PROJETO INTEGRADOR 1\\dados\\1st_2022"
second_path2022 = "C:\\Users\\SAMSUNG\\OneDrive\\Documentos\\PROJETO INTEGRADOR 1\\dados\\2nd_2022"

first_df_2022 = pd.read_csv(first_path2022, index_col=0)
second_df_2022 = pd.read_csv(first_path2022, index_col=0)
```

```python
''' 
PT-BR
    em analise explorativa, notou-se que algumas colunas
    mantiveram suas features, mas foram renomeadas a partir
    de 2023, gerando uma incoerências do ano de 2022 com os
    restantes:
     
    cidade de ocorrencia deixou de ser nomeada como "CIDADE"
    e passou a ser nomeada como "NOME_MUNICIPIO";

    data de registro deixou de ser nemada como
    "DATA_COMUNICACAO_BO" e passou a ser nomeada
    como "DATA_REGISTRO";

    descrição de periodo de ocorrencia deicou de ser nomeada
    como "DESCR_PERIODO" e passou a ser nomeada como
    "DESC_PERIODO".

    descrição do local de ocorrencia deixou de ser nomeada
    como "DESCR_TIPOLOCAL e passou a ser nomeada como
    "DESCR_SUBTIPOLOCAL".

EN
    in exploratory analysis, it was noted that some columns
    kept their features, but were renamed starting in 2023,
    creating inconsistencies between 2022 and the other years:

    the city of occurrence feature was no longer named "CIDADE"
    and was now named "NOME_MUNICIPIO";

    the registration date feature was no longer named
    "DATA_COMUNICACAO_BO" and was now named "DATA_REGISTRO";

    the description of the occurrence period feature was no longer
    named "DESCR_PERIODO" and was now named "DESC_PERIODO".

    the description of the occurrence location feature was no longer
    named "DESCR_TIPOLOCAL" and was now named
    "DESCR_SUBTIPOLOCAL".

'''

# renomando as colunas de 2022
# renaming 2022 columns

first_df_2022_renamed = first_df_2022.rename(columns={'CIDADE': 'NOME_MUNICIPIO',
                                                      'DATA_COMUNICACAO_BO': 'DATA_REGISTRO',
                                                      "DESCR_PERIODO": "DESC_PERIODO",
                                                      "DESCR_TIPOLOCAL": "DESCR_SUBTIPOLOCAL"})

second_df_2022_renamed = second_df_2022.rename(columns={'CIDADE': 'NOME_MUNICIPIO',
                                                      'DATA_COMUNICACAO_BO': 'DATA_REGISTRO',
                                                      "DESCR_PERIODO": "DESC_PERIODO",
                                                      "DESCR_TIPOLOCAL": "DESCR_SUBTIPOLOCAL"})
```

```python
# dados de 2023
# 2023 data

first_path2023 = "C:\\Users\\SAMSUNG\\OneDrive\\Documentos\\PROJETO INTEGRADOR 1\\dados\\1st_2023"
second_path2023 = "C:\\Users\\SAMSUNG\\OneDrive\\Documentos\\PROJETO INTEGRADOR 1\\dados\\2nd_2023"

first_df_2023 = pd.read_csv(first_path2023, index_col=0)
second_df_2023 = pd.read_csv(first_path2023, index_col=0)
```

```python
# dados de 2024
# 2024 data

first_path2024 = "C:\\Users\\SAMSUNG\\OneDrive\\Documentos\\PROJETO INTEGRADOR 1\\dados\\1st_2024"  
second_path2024 = "C:\\Users\\SAMSUNG\\OneDrive\\Documentos\\PROJETO INTEGRADOR 1\\dados\\2nd_2024"

first_df_2024 = pd.read_csv(first_path2024, index_col=0)
second_df_2024 = pd.read_csv(first_path2024, index_col=0)
```

# MERGE

```python
# junção dos dados
# data merging

list_df = [first_df_2022_renamed, second_df_2022_renamed, first_df_2023, second_df_2023, first_df_2024, second_df_2024]
df_merged = pd.concat(list_df, axis=0)
df_merged
```

```python
# normalizando "desc"
# normalizing "desc"

df = df_merged.rename(columns={'DESCR_SUBTIPOLOCAL': 'DESC_SUBTIPOLOCAL'})
df
```

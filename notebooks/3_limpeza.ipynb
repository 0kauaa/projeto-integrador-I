# limpeza

```python
# exluindo colunas desnescessárias para o estudo

del_columns = [
    'NOME_DEPARTAMENTO',
    'LATITUDE',
    'NOME_SECCIONAL',
    'LONGITUDE',
    'NOME_DELEGACIA',
    'NOME_DELEGACIA_CIRCUNSCRIÇÃO',
    'NUM_BO',
    'NOME_DEPARTAMENTO_CIRCUNSCRIÇÃO',
    'ANO_BO',
    'NOME_SECCIONAL_CIRCUNSCRIÇÃO',
    'LOGRADOURO',
    'NOME_MUNICIPIO_CIRCUNSCRIÇÃO',
    'NUMERO_LOGRADOURO',
    'DESCR_CONDUTA',
    'NATUREZA_APURADA'
]

df_col = df.drop(columns=[col for col in del_columns if col in df.columns])
df_col
```

```python
# filtrando os dados de santos

def city_filter(df, col, city):
    df[col] = df[col].str.strip().str.upper()

    return df[df[col] == city]

df_city = city_filter(df_col, "NOME_MUNICIPIO", "SANTOS")

df_city
```

```python
# filtrando ocorrencias de estupro

df_occ = df_city[(df_city['RUBRICA'].str.upper().str.contains("ESTUPRO", regex=False, na=False, case=False))]
df_occ
```

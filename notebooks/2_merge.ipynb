```python
# dados de 2022

first_path2022  = "/content/drive/My Drive/projeto-integrador-I/data/raw/1st_2022.xlsx"
second_path2022 = "/content/drive/My Drive/projeto-integrador-I/data/raw/2nd_2022.xlsx"

first_df_2022  = pd.read_csv(first_path2022, index_col=0)
second_df_2022 = pd.read_csv(first_path2022, index_col=0)
```

```python
''' 
    em analise explorativa, notou-se que algumas colunas
    mantiveram suas features, mas foram renomeadas a partir
    de 2023, gerando uma incoerências do ano de 2022 com os
    restantes:
     
    cidade de ocorrencia deixou de ser nomeada como "CIDADE"
    e passou a ser nomeada como "NOME_MUNICIPIO";

    data de registro deixou de ser nemada como
    "DATA_COMUNICACAO_BO" e passou a ser nomeada
    como "DATA_REGISTRO";

    descrição de periodo de ocorrencia deicou de ser nomeada
    como "DESCR_PERIODO" e passou a ser nomeada como
    "DESC_PERIODO".

    descrição do local de ocorrencia deixou de ser nomeada
    como "DESCR_TIPOLOCAL e passou a ser nomeada como
    "DESCR_SUBTIPOLOCAL".
'''

# renomando as colunas de 2022
# renaming 2022 columns

first_df_2022_renamed  = first_df_2022.rename(columns={'CIDADE': 'NOME_MUNICIPIO',
                                                      'DATA_COMUNICACAO_BO': 'DATA_REGISTRO',
                                                      "DESCR_PERIODO": "DESC_PERIODO",
                                                      "DESCR_TIPOLOCAL": "DESCR_SUBTIPOLOCAL"})

second_df_2022_renamed = second_df_2022.rename(columns={'CIDADE': 'NOME_MUNICIPIO',
                                                      'DATA_COMUNICACAO_BO': 'DATA_REGISTRO',
                                                      "DESCR_PERIODO": "DESC_PERIODO",
                                                      "DESCR_TIPOLOCAL": "DESCR_SUBTIPOLOCAL"})
```

```python
# dados de 2023

first_path2023  = "/content/drive/My Drive/projeto-integrador-I/data/raw/1st_2023.xlsx"
second_path2023 = "/content/drive/My Drive/projeto-integrador-I/data/raw/2nd_2023.xlsx"

first_df_2023  = pd.read_csv(first_path2023, index_col=0)
second_df_2023 = pd.read_csv(first_path2023, index_col=0)
```

```python
# dados de 2024

first_path2024  = "/content/drive/My Drive/projeto-integrador-I/data/raw/1st_2024.xlsx"
second_path2024 = "/content/drive/My Drive/projeto-integrador-I/data/raw/2nd_2024.xlsx"

first_df_2024  = pd.read_csv(first_path2024, index_col=0)
second_df_2024 = pd.read_csv(first_path2024, index_col=0)
```

```python
# dados de 2025

first_path2025  = "/content/drive/My Drive/projeto-integrador-I/data/raw/1st_2025.xlsx"
second_path2025 = "/content/drive/My Drive/projeto-integrador-I/data/raw/2nd_2025.xlsx"

first_df_2025  = pd.read_csv(first_path2025, index_col=0)
second_df_2025 = pd.read_csv(first_path2025, index_col=0)
```

```python
second_df_2025['DESCR_TIPOLOCAL'].unique()
```

```python
# o ano de 2025 tem 3 colunas a mais (cmd, ctl, cia)
df2025 = pd.concat([first_df_2025, second_df_2025], axis=0)
df2025 = df2025.drop(columns=['CMD', 'BTL', 'CIA', 'DESCR_SUBTIPOLOCAL'], axis=1) # colunas que só existem em 2025, remover para o merge
df2025 = df2025.rename(columns={
    'DESCR_TIPOLOCAL' : 'DESCR_SUBTIPOLOCAL' # renomear para manter o padrão adotado nos demais conjuntos
})

df2025.columns
```

```python
print(
       len(first_df_2022.columns) == len(second_df_2022.columns) 
    == len(first_df_2023.columns) == len(second_df_2023.columns) 
    == len(first_df_2024.columns) == len(second_df_2024.columns) 
    == len(df2025.columns)
    )
```

# MERGE

```python
# merge dos dados

list_df = [first_df_2022_renamed, second_df_2022_renamed, 
           first_df_2023, second_df_2023, 
           df2025]

df_merged = pd.concat(list_df, axis=0)
df_merged.tail()
```

```python
# normalizando "desc"

df = df_merged.rename(columns={'DESCR_SUBTIPOLOCAL': 'DESC_SUBTIPOLOCAL'})
df.columns
```

# tabela de frequencia

```python
# classes da tabela
k = int(np.ceil(1 +  3.3 * np.log10(len(delta)))) # regra de sturges
```

```python
# classes com amplitude variável
delta['classe'] = pd.qcut(delta['delta'], q=k, duplicates='drop')

# tabela inicial com frequencia de classe
freq = delta.groupby('classe', observed=False)['count'].sum().reset_index()
freq.columns = ['_intervalo', 'fi'] # _intervalo é apenas um suporte para os proximos calculos
freq.head()
```

```python
freq['intervalos'] = freq['_intervalo'].apply(
    lambda x: f"{x.left} |- {x.right}"
)

freq
```

```python
freq['xi'] = freq['_intervalo'].apply(
    lambda x: round(float((x.left + x.right) / 2), 1)
)

freq
```

```python
# produto medio
freq['xifi'] = freq['xi'].astype(float) * freq['fi'].astype(float)
freq
```

```python
# frequencia relativa(fr)
total = freq['fi'].sum()
freq['fr'] = freq['fi'] / total * 100
freq
```

```python
# frequencia acumulada absoluta (fac)
freq['fac'] = freq['fi'].cumsum()
freq
```

```python
# frequencia relativa acumulada (fri)
freq['fri'] = freq['fr'].cumsum()
freq
```

```python
# tablea final organizada
table = freq[['intervalos', 'fi', 'xi', 'xifi', 'fr', 'fac', 'fri']].copy()
table = table.reset_index(drop=True)
table
```

```python
table.to_excel('/content/drive/MyDrive/projeto-integrador-I/data/processed/freq_table.xlsx')
```

# estatisticas estimadas

```python
# media estimada
mean = int(table['xifi'].sum() / total)
mean
```

```python
# traduzindo para um formato de calendario
mean_mounth = int(mean / 30)
mean_days   = mean % 30

mean_mounth, mean_days
```

# tabela de frequencia da primeira semana

```python
# ocorrencias notificadas dentro da primeira semana (até 7 dias)
week = delta[delta['delta'] <= 7].sort_values(by='delta')
week = week.drop(columns='classe')
week = week.rename(columns={'delta' : 'classe'})
week
```

```python
# frequência de cada intervalo (fi)
week = week.rename(columns={'count' : 'fi'})
week
```

```python
# frequencia relativa (fr)
week_total = week['fi'].sum()
week['fr'] = week['fi'] / week_total * 100
week
```

```python
# frequencia acumulada absoluta (fac)
week['fac'] = week['fi'].cumsum()
week
```

```python
# frequencia relativa acumulada (fri)
week['fri'] = week['fr'].cumsum()
week
```

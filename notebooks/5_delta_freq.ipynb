# tabela de frequencia

```python
# parametros da tabela
rol = delta.sort_values(by='delta')
k = int(np.ceil(1 +  3.3 * np.log10(len(delta)))) # regra de sturges
```

```python
# classes com amplitude variável
delta['classe'] = pd.qcut(delta['delta'], q=k, duplicates='drop')

# tabela inicial com frequencia de classe
freq = delta.groupby('classe', observed=False)['count'].sum().reset_index()
freq.columns = ['_intervalo', 'fi'] # _intervalo é apenas um suporte para os proximos calculos
freq.head()
```

```python
# classes da tabela
freq['intervalos'] = freq['_intervalo'].apply(
    lambda x: f"{int(x.left)} |- {int(x.right)}"
)

freq
```

```python
# classes medias (xi)
freq['xi'] = freq['_intervalo'].apply(
    lambda x: float((x.left + x.right) / 2)
).round(1)

freq
```

```python
# produto medio
freq['xifi'] = freq['xi'].astype(float) * freq['fi'].astype(float)
freq
```

```python
# frequencia relativa(fr)
total = freq['fi'].sum()
freq['fr'] = freq['fi'] / total * 100
freq
```

```python
# frequencia acumulada absoluta (fac)
freq['fac'] = freq['fi'].cumsum()
freq
```

```python
# frequencia relativa acumulada (fri)
freq['fri'] = freq['fr'].cumsum()
freq
```

```python
# tablea final organizada
table = freq[['intervalos', 'fi', 'xi', 'xifi', 'fr', 'fac', 'fri']].copy()
table = table.reset_index(drop=True)
table
```

```python
table.to_excel('/content/drive/MyDrive/projeto-integrador-I/data/processed/freq_table.xlsx')
```
